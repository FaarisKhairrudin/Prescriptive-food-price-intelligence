```python
import pandas as pd
from playwright.async_api import async_playwright
```

```python
async def scrape_pihps_harian(
    start_date: str,
    end_date: str,
    provinsi: str = "Jawa Barat",
    kota: str = "Kota Bandung",
    komoditas: str = "Cabai Rawit Merah",
):
    """
    Scrape data harian PIHPS.

    Parameter tanggal:
    - start_date: 'YYYY-MM-DD'
    - end_date: 'YYYY-MM-DD'
    """

    url = "https://www.bi.go.id/hargapangan/TabelHarga/PasarTradisionalDaerah"

    async with async_playwright() as p:
        browser = await p.chromium.launch(
            headless=False  # False dulu buat lihat prosesnya saat testing
        )

        page = await browser.new_page()
        await page.goto(url, wait_until="networkidle")

        # Tunggu halaman filter muncul
        await page.wait_for_timeout(3000)

        # ======================================================
        # FILTER
        # ======================================================
        # NOTE:
        # Selector di bawah kemungkinan perlu disesuaikan.
        # Ini masih asumsi umum pola dropdown/input di halaman web.

        selects = page.locator("select")
        jumlah_select = await selects.count()
        print("Jumlah dropdown ditemukan:", jumlah_select)

        # Untuk debugging: lihat opsi select yang tersedia
        for i in range(jumlah_select):
            opsi = await selects.nth(i).locator("option").all_inner_texts()
            print(f"\nDropdown ke-{i}:")
            print(opsi[:20])

        # ======================================================
        # SEMENTARA STOP DI SINI UNTUK CEK STRUKTUR DROPDOWN
        # ======================================================

        await browser.close()
```

```python
await scrape_pihps_harian(
    start_date="2026-02-18",
    end_date="2026-05-20"
)
```
